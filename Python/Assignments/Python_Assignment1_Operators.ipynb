# Answer 3 : Operations on Int variable

# Answer 4: Boolean Operators 

```python


print (a, "==" ,b, "is", a == b )
print (a, "!=", b, "is", a != b )
print (a, ">", b, "is", a > b )
print (a, ">=", b, "is", a >= b )
print (a, "<", b, "is", a < b )
print (a, "<=", b, "is", a <= b )
```
