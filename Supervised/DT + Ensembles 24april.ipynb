##  Decision Tree Modelling

```python
# Separate the target and descriptive features
#X=employee.drop(['Attrition'],axis=1)     #dont do this first

y=employee['Attrition']
X=employee.drop(['Attrition'],axis=1)
print(X.shape)
```

```python
#Split the dataset into training and testing part
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)
print(y_train.head())
print(X_train.shape)
print(y_test.shape)
```

```python
# Step 3: Fit the model on the training dataset
#Import Tree Classifier model
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier()   #creating decision tree object 1 with default parameters gini index

#Train the model using the training sets
dt_model.fit(X_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = dt_model.predict(X_test)
print(y_pred)

#print(dt_model.tree_)
```

```python
print(dt_model.feature_importances_)
print(dt_model.max_features_)
print(dt_model.n_classes_)
print(dt_model.n_features_)
print(dt_model.classes_)
```

```python
#decision tree for model 1 with default parameters , spreaded to its maximum depth
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(40,20))
plot_tree(dt_model, class_names=['No','Yes'], feature_names=X.columns,filled=True)
plt.show()
```

* As the tree grows the features will become less significant because we are choosing the most important features to make the tree. So we can stop the tree at some stage by either prescribing maxdepth, or min_leaf_split i.e all parameters which help to prune the tree

```python
# Measure the performance of the model
#import scikit learn metrics to calculate accuracy
from sklearn import metrics
print ('Accuracy M1:',metrics.accuracy_score(y_test,y_pred))

#print("Precision:",metrics.precision_score(y_test, y_pred))
#print("Recall:",metrics.recall_score(y_test, y_pred)) 
```

*  Searching for the right value of hyperparameters using hit and trial in a real time problem may not be
easy. So define the range/list  of parameters we want to experiment with using a dictionary. Then give this
dictionary as input to the function GridSearchCV to identify the best combination of hyperparameters

```python
from sklearn.model_selection import GridSearchCV

# Define the parameters using a dictinary, use a full grid over all parameters
param_grid = {"max_depth": [2,3,4,5,6],
              "max_features": [1, 3, 5, 10],
              "min_samples_split": [2, 3, 5, 10],
              "criterion": ["gini", "entropy"], "min_samples_leaf":[1,2,3,4]}

# run grid search
grid_search = GridSearchCV(dt_model, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

#grid_search.best_params_  attribute gives the best parameters for our model
dt_params = grid_search.best_params_
print(dt_params)
```

```python
print(grid_search.best_params_)
dt_grid = DecisionTreeClassifier(**dt_params)

dt_grid.fit(X_train, y_train)

ypredg = dt_grid.predict(X_test)

print ('Accuracy:',metrics.accuracy_score(y_test,ypredg))
```

```python
print(metrics.classification_report(y_test, ypredg))
```

```python
#decision tree for model 3 with hyperparamters, spreaded to its maximum depth
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(40,20))
plot_tree(dt_grid, class_names=['No','Yes'], feature_names=X.columns, filled=True)
plt.show()
```

# Saving and using the model for prediction
* Python pickle module is used for serializing and de-serializing python object structures. 
* The process to converts any kind of python objects (list, dict, etc.) into byte streams (0s and 1s) is called pickling or serialization or flattening or marshalling.

```python
#saving the model 4 (grid model )into a pickle file

import pickle
model = 'final_model.pkl'      #name of the pickle file

#open the file in write mode
pickle.dump(dt_grid, open(model, 'wb'))  #saving the dt2 model using dump function
```

# Ensemble - Random Forest

```python
#import random forest 
from sklearn.ensemble import RandomForestClassifier

# build a classifier
rf = RandomForestClassifier()

#Train the model using the training sets
rf.fit(X_train, y_train)
```

```python
#Predict the response for test dataset
y_pred = rf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
# Model Precision

print(metrics.classification_report(y_test, y_pred))
```

# Random forest with GRID SEARCH

```python
from sklearn.model_selection import GridSearchCV

# use a full grid over all parameters
param_grid = {"max_depth": [3, None],
              "max_features": [1, 3, 10],
              "min_samples_split": [2, 3, 10],
              "bootstrap": [True, False],
              "criterion": ["gini", "entropy"]}

# run grid search
grid_search = GridSearchCV(rf, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

```python
dt_params = grid_search.best_params_
print(dt_params)
```

```python
# build a Random Forst classifier with best of Grid search
#rf_grid_search = RandomForestClassifier(n_estimators=100, bootstrap = False , criterion = 'entropy',  max_features = 10, min_samples_split = 10)
rf_grid_search = RandomForestClassifier(**dt_params)
#Train the model using the training sets with best of Grid search
rf_grid_search.fit(X_train, y_train)

#Predict the response for test dataset with best of Grid search
y_pred = rf_grid_search.predict(X_test)

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
