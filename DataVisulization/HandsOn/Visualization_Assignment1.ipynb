# Univariate Analysis
Q1 -  Plot Histograms using iris dataset

```python
figsize = (18,10)
plt.subplot(2,2,1)
sns.distplot(iris.sepal_length,bins=7)
plt.subplot(2,2,2)
sns.distplot(iris.sepal_width,bins=4)
plt.show()
plt.subplot(2,2,3)
sns.distplot(iris.petal_length)
plt.subplot(2,2,4)
sns.distplot(iris.petal_width)
plt.show()
```

Q2 - Plot Area Plot using iris dataset

```python
figsize = (16,10)
iris.plot.area(y=['sepal_length','sepal_width','petal_length','petal_width'],alpha=0.4);
plt.show()
```

Q3 Plot Densityplot using iris dataset

```python
figsize = (16,10)
plt.subplot(2,2,1)
sns.kdeplot(data=iris['sepal_length'], shade=True)
plt.subplot(2,2,2)
sns.kdeplot(data=iris['sepal_width'], shade=True)
plt.subplot(2,2,3)
sns.kdeplot(data=iris['petal_length'], shade=True)
plt.subplot(2,2,4)
sns.kdeplot(data=iris['petal_width'], shade=True)
plt.show()
```
