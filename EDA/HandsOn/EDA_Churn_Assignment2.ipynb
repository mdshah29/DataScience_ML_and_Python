```python
# removed non-corelated columns
telcom_processed = telcom.drop(['Account length','Area code','International plan','Voice mail plan','Total day calls','Total eve minutes','Total eve calls','Total night minutes','Total night calls','Total night charge'], axis = 1)
```

```python
#telcom_processed.groupby(['State'])['Churn'].sum()
# use catplot to split on Customer service calls 
sns.catplot(x='Churn', col='Customer service calls', kind='count', data=telcom_processed)   #factorplot()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (8, 6)
sns.countplot(x="Customer service calls", hue='Churn', data=telcom);
```

###### Observation:-
- we can see that Churn impacted for Customer service calls between 0 to 6 and it was max for Customer service call = 1
